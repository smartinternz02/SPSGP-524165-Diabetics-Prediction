Handle the Missing values.

```python
data.isnull().any()
```

```python
data.isna()
```

```python
data['age'].fillna(data['age'].mean(),inplace=True)
data['embarked'].fillna(value='unknown',inplace=True)
data['deck'].fillna(value='unknown',inplace=True)
data['embark_town'].fillna(value='unknown',inplace=True)
```

```python
data.isnull().any()
```

Find the outliers and replace the outliers

```python
sns.boxplot(x=data.age)
```

```python
perc99=data.age.quantile(0.99)
perc99
```

```python
data=data[data.age<=perc99]
sns.boxplot(data.age)
```

Check for Categorical columns and perform encoding.

```python
data.head()
```

```python
#label Encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
data.sex=le.fit_transform(data.sex)
data.embarked=le.fit_transform(data.embarked)
data.who=le.fit_transform(data.who)
data.adult_male=le.fit_transform(data.adult_male)
data.deck=le.fit_transform(data.deck)
data.alone=le.fit_transform(data.alone)
data.alive=le.fit_transform(data.alive)
```

```python
data.head()
```

```python
#onehot encoding 
data_main=pd.get_dummies(data,columns=['embark_town','class'])
```

```python
data_main.head()
```

 Split the data into dependent and independent variables.

```python
y=data_main['fare']
y.head()
```

```python
x=data_main.drop(columns=['fare'],axis=1)
x.head()
```

Scale the independent variables


```python
name=x.columns
name
```

```python
from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler()
```

```python
x_scale=scale.fit_transform(x)
x_scale
```

```python
x=pd.DataFrame(x_scale,columns=name)
x
```

Split the data into training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```
